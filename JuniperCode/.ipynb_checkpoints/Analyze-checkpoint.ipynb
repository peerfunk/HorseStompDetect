## Automated Detection Algorithm
The following code parts define an automated way of detecting the pattern in audio files.

### Sliding Window
The `SlidingWindow` - function takes a `Pattern`, a `SearchAble` - i.e.: the FFT of the AudioFile that is to be searched and the `WindowWidth`, which is defined as the length of the pattern. It goes over the whole `SearchAble`  and calculates the difference between the pattern and the current window this difference is then taken and used to calculate the similarity between them as a percentage - value. everything with a similarity > 95% is considered a siginificant find and classified as an alert-event.

```python
def getStartEnd(data,start , end):
    return np.array(data[:,list(range(start,end))])
def SlidingWindow(Pattern, SearchAble,WindowWidth):
    ret=[0]
    ret2=[]
    ret3=[]
    mean=0
    absmean=0
    last =0
    currentLength=0
    for sliding in range(1,len(SearchAble[1])-WindowWidth):
        Frame2=getStartEnd(SearchAble,sliding,sliding+WindowWidth)
        diff = Pattern - Frame2
        count=(diff<0).sum()
        cellCount = len(diff) * len(diff[0])
        perc = count/cellCount
        ret.append(perc)
        if perc > 0.95:
            ret2.append(sliding)
            ret3.append(1)
        else:
            ret3.append(0)
    return ret, ret2,ret3
```

### Filtering
In special edge cases which could only be found in synthetic tests, events where falsely classfied. The following filter-function is specific for this type of sound. Horses in stress situations paw with their hoofes in a spefic manner. Most of the time this is shown in multiple, quickly followed scratches, mostly followed by short pauses, in cases of extreme pain or stress pawing may not even have pauses. This quickly repeating pattern is considered a positive. Single events are seen as false-positives and filtered/left out.

```python
def filter(result,duration):
    counter=0
    ret=[]
    ret1=[]
    last=0
    durPerCell=duration/len(result)
    returnFrame=[]
    for x,i in enumerate(result):
        if i==1:
            if last == 0:
                ret.append(x)
            counter+=1
        else:
            if last==1:
                ret.append(x)
                counter =0
        last = i
    for j in range(0,len(ret)-2):
        space = ((ret[j+1]-ret[j])*durPerCell)
        if space < 10 and space > 0.6:
            #print(str(ret[j+1]-ret[j]) +"," + str(ret[j]))
            ret1.append((ret[j],(ret[j+1]-ret[j])))
    return ret1
```

### Load and Analyze a File
This function loads a file, analyzes it and filters the result. A graph is drawn to show the similarites of the loaded file and the Pattern over time. the second graph shows where the Pattern was detected in the file.

```python
def Analyze(File):
    y, sr = librosa.load(File)
    y = librosa.util.normalize(y)
    ipd.Audio(y, rate=sr)
    duration=librosa.get_duration(y=y, sr=sr)
    fft = np.abs(librosa.stft(y))**2
    result,result2,result3=SlidingWindow(Patterns,fft,len(Patterns[0]))
    ret=filter(result3,duration)
    result = np.multiply(100,result)
    part = (duration/60)/len(result)
    for retval in ret:
        print(retval)
        print("Minute: " + str(retval[0]*part))
    x1 = [i*part for i in range(len(result))]
    fig, ax = plt.subplots()
    plt.title('bereinstimmung des Samples mit dem Pattern')
    ax.plot(x1,result)
    ax.scatter([(i*part) for i in result2],[100]*len(result2),marker="D",color = 'red')
    fmtr = matplotlib.ticker.StrMethodFormatter('{x:2.2f}%')
    ax.yaxis.set_major_formatter(fmtr)
    plt.xlabel('Zeit (min)')
    plt.ylabel('Wahrscheinlichkeit')
    plt.grid()
    plt.show()
```
