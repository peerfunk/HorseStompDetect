```python
import numpy as np
import matplotlib.pyplot as plt
import librosa.display
import math
import matplotlib
import IPython.display as ipd
```

```python
x = np.linspace(np.pi, np.pi*20, 200)
func = np.sin(x)+np.sin(np.tan(x))
y = np.abs(librosa.stft(func))
```

```python
fig = plt.figure()
fig.suptitle('f(x)=sin(x)+sin(tan(x))', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)

ax.set_xlabel('Zeit')
ax.set_ylabel('Amplitude')

ax.plot(x, func)

plt.show()
```

```python
fig = plt.figure()
fig.suptitle('stft(f(x)=sin(x)+sin(tan(x)))', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)

ax.set_xlabel('Frequenz')
ax.set_ylabel('Frequenzdichte')

ax.plot(y)

plt.show()
```

```python
y, sr = librosa.load("..\\out1min.wav")
duration=librosa.get_duration(y=y, sr=sr)
fft = np.abs(librosa.stft(y))
part = (duration)/len(y)
x1 = [i*part for i in range(len(y))]
fig =plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x1, y)
plt.title('Darstellung einer Test-Audiodatei')
plt.xlabel('Zeit (s)')
plt.ylabel('Amplitude')


plt.subplot(2, 1, 2)
plt.title('STFT einer Test-Audiodatei')


librosa.display.specshow(librosa.amplitude_to_db(fft,ref=np.max), cmap='gray_r', x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')

fig.tight_layout()
plt.show()
```
